## Simple Line Plots

The visualization of a single function *y = f(x)*.

```python
plt.style.use('seaborn-v0_8-bright')

fig = plt.figure()  # Creating a figure
ax = plt.axes()  # Creating an axes

x = np.linspace(0,10,1000)
ax.plot(x, np.sin(x)*10)
ax.plot(x, np.cos(x)*5)
```

```python
plt.style.use('seaborn-v0_8-bright')
fig = plt.figure()
ax = plt.axes()

x = np.linspace(0,10,100)

ax.plot(x, np.sin(x)*10, label='sin(x)*10', color='skyblue')
ax.plot(x, np.cos(x)*5, label='cos(x)*5', color='red')
ax.set_xlabel('x')
ax.set_ylabel('Trignometric function of x')
ax.set_title('A simple line chart')
ax.set_xlim(0,10)
ax.set_ylim(-10,10)
ax.set_xticks(np.arange(0,11))
```

## Scatter Plots

Close cousin of the line plot, the points are represented individually with a dot, circle, or other shape.

```python
plt.style.use('seaborn-v0_8-whitegrid')
x = 4 + np.random.normal(0, 2, 24)
y = 4 + np.random.normal(0, 2, len(x))
sizes = np.random.uniform(15, 80, len(x))
colors = np.random.uniform(15, 80, len(x))
# plot
fig, ax = plt.subplots()

ax.scatter(x, y, s=sizes, c=colors, vmin=0, vmax=100)

ax.set_xlim(0,8)
ax.set_ylim(0,8)
ax.set_xticks(np.arange(0,8))
ax.set_yticks(np.arange(0,8))
```

```python
plt.style.use('_mpl-gallery')
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
fig, ax = plt.subplots()
colors = rng.randn(100)
sizes = 100*rng.randn(100)
ax.scatter(x,y, c=colors, s=sizes, alpha=0.3)
```
