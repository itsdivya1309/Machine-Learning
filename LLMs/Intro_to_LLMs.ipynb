# Using Phi-3-mini

It is relatively small (3.8 B parameters) but a performant model.

When using a LLM, two models are loaded:
* the generative model itself
* its underlying tokenizer

```python
from transformers import AutoModelForCausalLM, AutoTokenizer

# Load the model
model = AutoModelForCausalLM.from_pretrained(
    "microsoft/Phi-3-mini-4k-instruct",
    device_map="cuda",
    torch_dtype="auto",
    trust_remote_code=True
)
# Load the tokenizer
tokenizer = AutoTokenizer.from_pretrained("microsoft/Phi-3-mini-4k-instruct")
```

```python
from transformers import pipeline

# Create a pipleline
generator = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    return_full_text=False,
    max_new_tokens=500,
    do_sample=False
)
```

```python
# The prompt (user input/query)
message = [
    {"role":"user", "content":"Write a short poem like Thomas Hardy."}
]
# Generate output
output = generator(message)
print(output[0]['generated_text'])
```
