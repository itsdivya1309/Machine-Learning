## Simulating a Random Walk Process

A *random walk* is a process in which there is an equal chance of going up or down by a random number. In a random walk process, we say that the present value `yt` is a function of the value at the previous timestep `yt1`, a constant `C`, and a random number `t`, also termed white noise.

`yt = C + yt1 + t`

```python
rng = np.random.default_rng()

# Generate 1000 random numbers from a normal distribution with a mean 0 and standard deviation 0
steps = rng.standard_normal(1000)
# Initialize the first value of the series to 0
steps[0] = 0

# Calculate the cumulative sum of errors for each timestep in the simulated process.
random_walk = np.cumsum(steps)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(random_walk, color ='m')
plt.title('Simulating a Random Walk Process of 1000 time steps')
plt.xlabel('Timesteps')
plt.ylabel('Values')
```

## Identifying a Random Walk

In the context of time series, a random walk is a series whose first difference is stationary and uncorrelated.

***Steps to identify Random Walk***:
1. Check for stationarity: if the series is non-stationary, apply transformations such as differencing or log to make the series stationary.
2. Stationarity of a series can be checked by statistical tests such as the **Augmented Dicky Fuller Test**.
3. Once we have stationary series, we plot the **autocorrelation fucntion (ACF)**.

```python
ADF_result = adfuller(random_walk)
ADF_result
```

```python
print(f'ADF Statistic: {ADF_result[0]}')
print(f'p-value: {ADF_result[1]}')
```

The ADF Statistic score is not a very negative number, and the p-value is larger than 0.05. Hence, we fail to reject our null-hypothesis and the series is non-stationary. Next we plot the ACF.

```python
plot_acf(random_walk, lags=40)
plt.show()
```

The autocorrelation of values decreases with an increase in time lags. This again proves that our series is non-stationary. Now, we need to make our series stationary.

```python
first_order_diff = np.diff(random_walk, n=1)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(first_order_diff, color ='orange')
plt.title('First order differencing of our random walk')
plt.xlabel('Timesteps')
plt.ylabel('Values')
```

```python
# Again testing stationarity
adf_result_diff = adfuller(first_order_diff)
print(f'ADF Statistic: {adf_result_diff[0]}')
print(f'p-value: {adf_result_diff[1]}')
```

```python
plot_acf(first_order_diff, lags=20)
plt.show()
```

We have demonstrated that our simulated data is indeed a random walk: it is stationary and uncorrelated after first order differencing.

## Is GOOGL Stock data a random walk?

```python
# AD Fuller test of the series

adf_result = adfuller(df['Close'])
print(f'AD Fuller Statistic Score: {adf_result[0]}')
print(f'p-value: {adf_result[1]}')
```

```python
plot_acf(df['Close'])
plt.show()
```

```python
# First order differencing
first_order_diff = list(np.diff(df['Close'], n=1))
first_order_diff.insert(0,0)
df.loc[:,'first_order_diff'] = first_order_diff
df
```

```python
# Stationarity test on first order difference
diff_result = adfuller(df['first_order_diff'])
print(f'A D Fuller Statistic Score: {diff_result[0]}')
print(f'p-value: {diff_result[1]}')
```

```python
# ACF Plot of first order difference
plot_acf(df['first_order_diff'], lags=20)
plt.show()
```

Hence, we conclude that the closing price of GOOGL can be approximated as a random process.

## Forecasting a Random Walk (simulated data)

```python
random_walk_df = pd.DataFrame({'Value': random_walk})
random_walk_df.head()
```

```python
train = random_walk_df.iloc[:800]
test = random_walk_df.iloc[800:]
```

```python
train
```

```python
plt.figure(figsize=(10, 6))
plt.plot(train['Value'], 'g-', label='Train')
plt.plot(test['Value'], 'g-', label='Test')
plt.xlabel('Time steps')
plt.ylabel('Value')
plt.axvspan(800, 1000, color='#808080', alpha=0.2)
plt.legend(loc=2)

plt.tight_layout()
```

```python
# Predicting the historical mean
mean = np.mean(train)
test.loc[:,'pred_mean'] = mean
test.head()
```

```python
# Predicting the last value
last = train['Value'].iloc[-1]
test.loc[:,'pred_last'] = last
test.head()
```

```python
# Drift method
deltaX = 800-1
deltaY = last - 0
drift = deltaY / deltaX
print(drift)
```

```python
x_vals = np.arange(800, 1000, 1)
pred_drift = x_vals * drift
test.loc[:, 'pred_drift'] = pred_drift
test.head()
```

```python
# Plot all the three forecasts
plt.figure(figsize=(10, 6))

plt.plot(train['Value'], 'g-', label='Train')
plt.plot(test['Value'], 'g-', label='Test')
plt.plot(test['pred_mean'], 'r-', label='Mean')
plt.plot(test['pred_last'], 'b-.', label='Last')
plt.plot(test['pred_drift'], 'k:', label='Drift')
plt.xlabel('Time steps')
plt.ylabel('Value')
plt.axvspan(800, 1000, color='#808080', alpha=0.2)
plt.legend(loc=2)

plt.tight_layout()
```

```python
mse_mean = mean_squared_error(test['Value'], test['pred_mean'])
mse_last = mean_squared_error(test['Value'], test['pred_last'])
mse_drift = mean_squared_error(test['Value'], test['pred_drift'])
print(mse_mean, mse_last, mse_drift)
```
