# Autoregressive Processes

## Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Dataset

Here, we want to predict the average weekly foot traffic in a retail store using past data. We have 1000 data points, each representing the average weekly foot traffic at a retail store since the year 2000.

```python
df = pd.read_csv('datasets/foot_traffic.csv')
df.head()
```

```python
fig, ax = plt.subplots()

ax.plot(df['foot_traffic'])
ax.set_xlabel('Time')
ax.set_ylabel('Average weekly foot traffic')

plt.xticks(np.arange(0, 1000, 104), np.arange(2000, 2020, 2))

fig.autofmt_xdate()
plt.tight_layout()
```
